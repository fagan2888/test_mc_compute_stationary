## mpmath

mc_compute_stationary_mpmath uses `mpmath.eig` from [mpmath](http://mpmath.org/).

```python
import sympy.mpmath as mp
```

```python
mp.__version__
```

The working precision is controlled by a context object called `mp`.  
See: [Setting the precision](http://mpmath.org/doc/current/basics.html#setting-the-precision)

```python
print mp.mp
```

The routine `eig` solves the eigenvalue problem.  
See: [The eigenvalue problem](http://mpmath.org/doc/current/matrices.html#the-eigenvalue-problem)

```python
E_A, EL_A = mp.eig(mp.matrix(A), left=True, right=False)
E_A, EL_A[0, :]/sum(EL_A[0, :]), EL_A[1, :]/sum(EL_A[1, :])
```

Let us see what happens if we increase the precision.  
See: [Setting the precision](http://mpmath.org/doc/current/basics.html#setting-the-precision)

```python
mp.mp.dps += 15
```

```python
E_A, EL_A = mp.eig(mp.matrix(A), left=True, right=False)
E_A, EL_A[0, :]/sum(EL_A[0, :]), EL_A[1, :]/sum(EL_A[1, :])
```

Arithemtic seems to be more accurate if the `mpf` values are created from strings.  
See: [Providing correct input](http://mpmath.org/doc/current/basics.html#providing-correct-input)

```python
AA = np.array(A).astype('str')
AA
```

```python
E_AA, EL_AA = mp.eig(mp.matrix(AA), left=True, right=False)
E_AA, EL_AA
```

```python
EL_AA[1, :]/sum(EL_AA[1, :])
```
